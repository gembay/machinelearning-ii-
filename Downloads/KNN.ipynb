```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
#LOADING THE DATE INTO A PANDAS DDATAFRAME
```

```python
data = pd.read_csv("C:/Users/USER/Desktop/archive (16)/train.csv")
```

```python
data.head()
```

```python
data["Pclass"].unique()
```

```python
#PREPROCESING OF THE DATA
```

```python
columns_to_drop = ["PassengerId","Name","Ticket","Cabin","Embarked"]
```

```python
data.drop(columns=columns_to_drop,inplace=True)
```

```python
data.head()
```

```python
##Normalize or scaling the features
```

```python
numerical_columns = ['Age','SibSp','Parch','Fare']
```

```python
data[numerical_columns]=(data[numerical_columns]-data[numerical_columns].min())/(data[numerical_columns].max()-data[numerical_columns].min())
```

```python
data.head()
```

```python
#NOW WE LABEL ENCODE
```

```python
data['Sex'].unique()
```

```python
label_encoder=LabelEncoder()
```

```python
data['Sex'] = label_encoder.fit_transform(data['Sex'])
```

```python
data.head()
```

```python
mean_age = data['Age'].mean()

# Fill the missing values in the 'age' column with the mean
data['Age'] = data['Age'].fillna(mean_age)
```

```python
#SPLIT THE INPUT AND THE TARGET
```

```python
x=data.drop("Survived",axis=1)
y=data["Survived"]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
#TRAINING THE MODEL
```

```python
k = 5
knn = KNeighborsClassifier(n_neighbors = k)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```
